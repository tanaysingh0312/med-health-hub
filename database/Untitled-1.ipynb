```python
board = ['1','2','3','4','5','6','7','8','9']
turn = 'X'

def display_board():
    print()
    for i in range(0, 9, 3):
        print(" | ".join(board[i:i+3]))
        if i < 6:
            print("--|---|--")

def check_win():
    combos = [
        [0,1,2], [3,4,5], [6,7,8],
        [0,3,6], [1,4,7], [2,5,8],
        [0,4,8], [2,4,6]
    ]
    for combo in combos:
        if board[combo[0]] == board[combo[1]] == board[combo[2]]:
            return True
    return False

def is_draw():
    return all(c in ['X','O'] for c in board)

display_board()
while True:
    move = int(input(f"\nPlayer {turn}, choose cell (1-9): ")) - 1
    if board[move] in ['X', 'O']:
        print("Cell already taken! Try again.")
        continue

    board[move] = turn
    display_board()

    if check_win():
        print(f"\nPlayer {turn} wins!")
        break
    if is_draw():
        print("\nIt's a draw!")
        break

    turn = 'O' if turn == 'X' else 'X'
```
