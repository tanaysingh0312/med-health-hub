```python
from collections import deque

def water_jug(jug1, jug2, target):
    visited = set()
    queue = deque([(0, 0, 0)])  # (x, y, steps)

    while queue:
        x, y, steps = queue.popleft()

        if (x, y) in visited:
            continue
        visited.add((x, y))

        print(f"({x}, {y})")

        if x == target or y == target:
            print(f"Reached target in {steps} steps!")
            return

        # All possible actions
        queue.append((jug1, y, steps+1))  # Fill jug1
        queue.append((x, jug2, steps+1))  # Fill jug2
        queue.append((0, y, steps+1))     # Empty jug1
        queue.append((x, 0, steps+1))     # Empty jug2

        # Pour jug1 → jug2
        pour = min(x, jug2 - y)
        queue.append((x - pour, y + pour, steps+1))

        # Pour jug2 → jug1
        pour = min(y, jug1 - x)
        queue.append((x + pour, y - pour, steps+1))

    print("No solution found.")

jug1, jug2, target = 4, 3, 2
water_jug(jug1, jug2, target)
```
